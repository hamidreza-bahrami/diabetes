```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
diabete = pd.read_csv('C:\\Users\\lenovo\\Desktop\\diabetes\\diabetes.csv')
diabete
```

```python
y = diabete.drop(['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'], axis=1)
y
```

```python
x = diabete.drop(['Outcome'], axis=1)
x
```

```python
x = np.array(x)
x
```

```python
y = diabete['Outcome'].to_numpy()
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 50)
model = RandomForestClassifier(n_estimators = 50, random_state = 50)
model.fit(x_train, y_train)
y_prediction = model.predict(x_test)
y_prediction
```

```python
import pickle
```

```python
data = {'model': model, 'x': x}
with open('saved.pkl', 'wb') as file:
    pickle.dump(data, file)
```
